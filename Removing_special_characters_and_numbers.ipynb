```python
import re
import emoji
def remove_special(input_string):
  pattern='[^A-Za-z ]'
  cleaned_string=re.sub(pattern, '',input_string)
  cleaned_string=emoji.demojize(cleaned_string)
  cleaned_string=cleaned_string.split()
  cleaned_string=' '.join(cleaned_string)
  return cleaned_string
# input_string = "This is a string with special characters: !@#$%^&*() and numbers 12345. 😊🚀"
# remove_special(input_string)
```

```python
df['Cleaned_summary']=df['Summary'].apply(lambda x:str(x))
df['Cleaned_summary']=df['Cleaned_summary'].apply(lambda x:remove_special(x))
df
```
